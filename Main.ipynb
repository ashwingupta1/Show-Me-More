```python
shows = shows[["id","name","overview","homepage", "poster_path","genres", "created_by"]]
```

```python
shows.dropna(subset=['poster_path'], inplace=True)
```

```python
shows.head(10)
```

```python
shows = shows.copy()
shows['tags'] =  shows['overview'] +" "+ shows['genres'] +" "+ shows['created_by']
```

```python
shows
```

```python
new_data = shows.drop(columns = ["id","overview", "genres", "created_by"])
```

```python
new_data.reset_index(drop=True, inplace=True)
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=8007, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[2]
```

```python
new_data[new_data['name'] =="Stranger Things"]. index[0]
```

```python
distance = sorted(list(enumerate(similarity[2])),reverse = True, key=lambda vector: vector[1])
for i in distance[1:6]:
    print (new_data.iloc[i[0]]["name"])
```

```python
def recommend(show):
    index = new_data[new_data['name'] ==show]. index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse = True, key=lambda vector: vector[1])
    for i in distance[1:6]:
        print (new_data.iloc[i[0]]["name"])
```

```python
recommend("Breaking Bad")
```
